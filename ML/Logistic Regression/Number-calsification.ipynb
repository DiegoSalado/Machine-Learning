Importamos las librerias, metodos y base de datos necesaria

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
%matplotlib inline
digits=load_digits()
```

Determinamos el numero de images

```python
print('imagen data shape: ',digits.data.shape)
print('label data shape: ', digits.target.shape)
```

Dysplaying some of the images and labels

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(20,4))
for index, (image, label) in enumerate(zip(digits.data[0:5],digits.target[0:5])):
    plt.subplot(1,5,index+1)
    plt.imshow(np.reshape(image,(8,8)),cmap=plt.cm.gray)
    plt.title('Training: %i\n' % label)
```

Dividing dataset into Training and Test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(digits.data,digits.target, test_size=0.23, random_state=2)
```

Import the Logistic Regression model and making an instance

of the model and training it

```python
from sklearn.linear_model import LogisticRegression

logisticRegr=LogisticRegression()
logisticRegr.fit(x_train,y_train)
```

Predicting the outpou of the first element of the test set

```python
print(logisticRegr.predict(x_test[0].reshape(1, -1)))
```

Predicting the output of the first 10 elements of the test set

```python
logisticRegr.predict(x_test[0:10])
```

Determining the accuaracy of the model

```python
predictions=logisticRegr.predict(x_test)

score=logisticRegr.score(x_test,y_test)
print(score)
```

Representing the confusion matrix in a heatmap

```python
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test,predictions)
plt.figure(figsize=(9,9))
sns.heatmap(cm,annot=True,fmt='.3f',cmap='Blues_r', square=True)
plt.ylabel('Label', fontsize=17)
plt.xlabel('Predicted label', fontsize=17)
plt.title('Accuracy Score', fontsize=20)
```

Showing some images whit their predictions

```python
plt.figure(figsize=(20,3))
for index,(predict,actual) in enumerate(zip(predictions[0:5],y_test[0:5])):
    plt.subplot(1,5,index+1)
    plt.imshow(np.reshape(x_test[index],(8,8)), cmap=plt.cm.gray)
    plt.title('Predicted: {}, Actual: {}'.format(predict,actual),fontsize=17)
```
